```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/Data[1].csv")
```

```python
print("missing value in df \ n",df.isnull().sum())
```

```python
#fill with constants
df_fill_const=df.fillna("unknown")
print("after filling missng values values\n",df_fill_const.head())
```

```python
#fill with numeric values
df_fill_mean=df.fillna(df.mean(numeric_only=True))
print("After filling missing value\n",df_fill_mean)
```

```python
df_fill_mean.describe()
```

```python
#Dropping values
df_drop=df.dropna()
print("After deleting missing values\n",df_drop.isnull().sum())
```
